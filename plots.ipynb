```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

ts = pd.Series(np.random.randn(1000), index = pd.date_range('1,1,2000', periods = 1000))
```

```python
ts = ts.cumsum()
ts.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1,1,2000', periods = 1000))
df = pd.DataFrame(np.random.randn(1000,4), index = ts.index, columns = list('ABCD'))
df = df.cumsum()
plt.figure()
df.plot()
plt.show()
```

```python
df3 = pd.DataFrame(np.random.randn(1000,2), columns = ['B','C']).cumsum()
df3['A'] = pd.Series(list(range(len(df))))
df3.iloc[5].plot.bar()
plt.axhline(0,color = 'R')
plt.show()
```

```python
df4 = pd.DataFrame({'a':np.random.randn(1000)+1, 'b':np.random.randn(1000), 'c':np.random.randn(1000)-1}, columns = ['a','b','c'])
plt.figure()
df4.plot.hist(alpha = 0.5)
plt.show()
```

```python
df5 = pd.DataFrame(np.random.rand(10,5), columns = ['A','B','C','D', 'E'])
df5.plot.box()
plt.show()
```

```python
df5.plot.area()
```

```python
df6 = pd.DataFrame(np.random.rand(500,4), columns = ['a','b','c','d'])
df6.plot.scatter(x='a', y='b')
plt.show()
```

```python
df8 = pd.DataFrame(np.random.randn(1000,2), columns = ['a','b'])
df8['a'] = df8['a'] + np.arange(1000)
df8.plot.hexbin(x='a', y='b', gridsize = 25)
plt.show()
```
